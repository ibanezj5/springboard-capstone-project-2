```python
import pandas as pd
import numpy as np
import math as ma


market_data = pd.read_csv("data/marketdata_sample.csv")
news_data = pd.read_csv("data/news_sample.csv")
```

```python
news_copy = news_data
news_copy['sentiment'] = ( news_copy['sentimentPositive'] - news_copy['sentimentNegative'] )*100
sentiment_figure_array = np.empty(100)
sentiment_array = np.array(news_copy['sentiment'])
index = 0

for i in news_copy['sentimentClass']:             
    if i == 0:
        sentiment_figure_array[index] = 1
    else:
        sentiment_figure_array[index] = sentiment_array[index]
    index += 1

news_copy['sentimentFigure'] = sentiment_figure_array
news_copy['relevanceFigure'] = news_copy['relevance']*10

```

```python
audiences_figure_array = np.empty(100)
i = 0
for record in news_copy['audiences']:
    audiences_figure_array[i] = ma.exp(len(record)/10)
    i += 1

news_copy['audiencesFigure'] = audiences_figure_array
print(news_copy[['audiences','audiencesFigure']])
#len(news_data['audiences'][19])
```

```python
relevance_figure_array = news_copy['relevanceFigure']
sentiment_figure_array = news_copy['sentimentFigure']
audiences_figure_array = news_copy['audiencesFigure']
```

```python
#media influence 12h feature engineering

volume_counts_12h_array = news_copy['volumeCounts12H']
news_copy['noveltyCount12H_mod'] = news_copy['noveltyCount12H']
novelty_count_12h_array = news_copy['noveltyCount12H_mod']
media_influence_12h_array = np.empty(100)

for i in range(100):
    novelty_count_12h_array[i] += 1

for i in range(100):
        media_influence_12h_array[i] = volume_counts_12h_array[i]*novelty_count_12h_array[i]*relevance_figure_array[i]*sentiment_figure_array[i]*audiences_figure_array[i]
    
news_copy['mediaInfluence12H'] = media_influence_12h_array
news_copy[['mediaInfluence12H', 'noveltyCount12H']] 
```
